```python
from sklearn.model_selection import train_test_split

def split_dataframe(df, train_size=0.6, val_size=0.2, test_size=0.2):
    """
    Splits a DataFrame into training, validation, and test sets.

    Parameters:
    - df: DataFrame to split.
    - train_size: Proportion of the dataset to include in the train set.
    - val_size: Proportion of the dataset to include in the validation set.
    - test_size: Proportion of the dataset to include in the test set. Should be equal to 1 - train_size - val_size.

    Returns:
    - df_train: Training set DataFrame.
    - df_val: Validation set DataFrame.
    - df_test: Test set DataFrame.
    """

    assert train_size + val_size + test_size == 1, "The sum of train_size, val_size, and test_size must be 1."

    # First, split the DataFrame into training and temp (validation + test) sets
    df_train, df_temp = train_test_split(df, test_size=val_size+test_size, random_state=42)

    # Next, split the temp set into validation and test sets
    df_val, df_test = train_test_split(df_temp, test_size=test_size/(test_size+val_size), random_state=42)

    return df_train, df_val, df_test


df_1, df_3, df_2 = split_dataframe(df, train_size=0.6, val_size=0.2, test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from nltk.translate.bleu_score import sentence_bleu

def compute_metrics(pred):
    references = pred.label_ids
    generated_texts = pred.predictions

    bleu_scores = []
    for reference, generated_text in zip(references, generated_texts):
        reference_text = data[reference]['answertext']
        bleu_score = sentence_bleu([reference_text], generated_text)
        bleu_scores.append(bleu_score)

    return {
        'bleu': sum(bleu_scores) / len(bleu_scores)
    }
```

gradient_accumulation_steps=4:

This specifies how many forward passes the model makes before a backward pass is performed to update the weights.
It allows you to effectively increase the batch size, especially when the hardware can't accommodate large batches due to memory limitations.
For instance, if you have a real batch size of 1 (as you do in your per_device_train_batch_size=1) but you set gradient_accumulation_steps to 4, it means the model will accumulate gradients over 4 forward passes before performing a single update. This mimics a larger batch size of 4.
warmup_steps=2:

This is a hyperparameter for the learning rate schedule.
During the initial warmup_steps, the learning rate gradually increases from a low value (or zero) to the set learning rate.
This gradual warm-up can help in stabilizing the training process and leads to better convergence, especially at the start of training.
max_steps=100:

This sets the total number of training steps to perform before stopping training.
A step is a single gradient update; if you have a dataset of 1000 samples and a batch size of 10, it would take 100 steps to complete one epoch.
When max_steps is set, training will stop after this number of steps regardless of how many epochs have been completed. The difference between num_train_epochs and the other parameters is that num_train_epochs controls training duration based on the dataset size, while the others control how the training process happens at each step or how learning rate changes with time. You can use max_steps as an alternative to num_train_epochs when you want to specify the exact number of updates to perform, which can be particularly useful for fine-tuning or when you want to ensure that training lasts for a specific number of iterations regardless of the size of the dataset.

```python


def formatting_func(example):
    text = f"Instruction: {example['questionText'][0]}\Response: {example['answerText'][0]}"
    return [text]

trainer = SFTTrainer(
    model=model,
    train_dataset=data,
    eval_dataset=data1,
        compute_metrics=compute_metrics,
    args=transformers.TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        warmup_steps=2,
        max_steps=100,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=1,
        output_dir="outputs",
        optim="paged_adamw_8bit"

    ),
    peft_config=lora_config,
    formatting_func=formatting_func,
    max_seq_length=8192
)
trainer.train()
#trainer.evaluate()
```

```python
import pandas as pd
from nltk.translate.bleu_score import sentence_bleu
from rouge_score import rouge_scorer

# Assuming df is your dataframe and it has 'answerText' and 'answer' columns
# df = pd.DataFrame({'answerText': [...], 'answer': [...]})

# Function to calculate BLEU score for a single pair of texts
def calculate_bleu(reference, candidate):
    return sentence_bleu([reference.split()], candidate.split())

# Function to calculate ROUGE scores for a single pair of texts
def calculate_rouge(reference, candidate):
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores = scorer.score(reference, candidate)
    return scores

# Apply the BLEU score calculation to each row in the DataFrame
df_2['bleu_score'] = df_2.apply(lambda row: calculate_bleu(row['answerText'], row['answer']), axis=1)

# Apply the ROUGE score calculation to each row in the DataFrame
df_2['rouge_scores'] = df_2.apply(lambda row: calculate_rouge(row['answerText'], row['answer']), axis=1)

# Display the DataFrame with the new columns for scores
print(df_2[['bleu_score', 'rouge_scores']])
```

```python
# Average BLEU score
average_bleu_score = df_2['bleu_score'].mean()
print(f"Average BLEU score: {average_bleu_score}")

# Average ROUGE scores
# Initialize sums for each ROUGE metric
total_rouge1 = 0
total_rouge2 = 0
total_rougeL = 0

# Loop through the DataFrame to sum up ROUGE scores
for index, row in df_2.iterrows():
    total_rouge1 += row['rouge_scores']['rouge1'].fmeasure
    total_rouge2 += row['rouge_scores']['rouge2'].fmeasure
    total_rougeL += row['rouge_scores']['rougeL'].fmeasure

# Calculate the average for each ROUGE score
average_rouge1 = total_rouge1 / len(df_2)
average_rouge2 = total_rouge2 / len(df_2)
average_rougeL = total_rougeL / len(df_2)

# Print the average ROUGE scores
print(f"Average ROUGE-1 F1 score: {average_rouge1}")
print(f"Average ROUGE-2 F1 score: {average_rouge2}")
print(f"Average ROUGE-L F1 score: {average_rougeL}")
```

```python
import pandas as pd
from bert_score import score

# Assuming df is your dataframe with 'answerText' as reference and 'answer' as generated columns
# For example:
# df = pd.DataFrame({
#     'answerText': ['This is the first reference sentence.', 'This is the second reference sentence.'],
#     'answer': ['This is the first generated sentence.', 'This is the second generated sentence.']
# })

# Convert DataFrame columns to lists for processing with BERTScore
references = df_2['answerText'].tolist()
candidates = df_2['answer'].tolist()

# Calculate BERTScore
P, R, F1 = score(candidates, references, lang="en", verbose=True)

# Adding BERTScore to the DataFrame
df_2['bertscore_precision'] = P.numpy()
df_2['bertscore_recall'] = R.numpy()
df_2['bertscore_f1'] = F1.numpy()
# Calculate the average BERTScore for the entire DataFrame
average_precision = df_2['bertscore_precision'].mean()
average_recall = df_2['bertscore_recall'].mean()
average_f1 = df_2['bertscore_f1'].mean()

# Print the average scores
print(f"Average BERTScore Precision: {average_precision}")
print(f"Average BERTScore Recall: {average_recall}")
print(f"Average BERTScore F1: {average_f1}")


```
